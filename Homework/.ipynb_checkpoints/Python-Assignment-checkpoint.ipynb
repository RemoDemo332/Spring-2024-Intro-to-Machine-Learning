### Math Refresher

For the problems below, I'll remind you of some ideas from Calculus.

* The $n$th degree **Taylor polynomial** an $n$-times differentiable function $f$ centered at $x=c$ is defined to be

$$P_{n,c,f}(x)=\sum\limits_{j=0}^n \frac{f^{(j)}(c)}{j!}(x-c)^j$$

* **Taylor's Theorem** says, for some $t\in(c,x)$,

$$f(x) = P_{n,c,f}(x) + \frac{f^{(n+1)}(t)}{n!}(x-t)^n(x-c)$$

* This means the error between the original function $f$ and its $n$th degree Taylor polynomial $P_{n,c,f}$ is less than or equal to the maximum value the **remainder term** can take: $\max\limits_{t\in(c,x)}\left|\frac{f^{(n+1)}(t)}{n!}(x-t)^n(x-c)\right|$

```python
import math 

''' Problem #2'''
''' Calculating for e^x '''
def taylor_poly(func, func_n, n, c, x):
    output = func(c)
    
    for i in range (1, n+1):
        output += func_n(c, i) * ((x-c) ** i) / math.factorial(i)
    
    return output

def e_power(x):
    return math.exp(x)

def n_e_power(x, n):
    return math.exp(x)

print('output:', taylor_poly(e_power, n_e_power, 6, 0, 1), '\n', 'x=6 n=1')

print('output:', taylor_poly(e_power, n_e_power, 5, 0, 0.5), '\n', 'x=5 n=0.5')

''' Calculating for 1 / 1 - x'''
def geo_equation(x):
    return 1 / (1 - x)

def n_geo_equation(x, n):
    return math.factorial(n) / (1 - x)**(n + 1)

print('output:', taylor_poly(geo_equation, n_geo_equation, 6, 0, 0.8), '\n', 'x=6 n=0.8')

print('output:', taylor_poly(geo_equation, n_geo_equation, 5, 0, 0.5), '\n', 'x=5 n=0.5')

''' Calculating for sin(x - pi/2) '''

def sin_equation(x):
    return math.sin(x - math.pi/2)

def n_sin_equation(x, n):
    if n % 4 == 0:
        return -math.cos(x)
    elif n % 4 == 1:
        return -math.sin(x)
    elif n % 4 == 2:
        return math.cos(x)
    else:  # n % 4 == 3
        return math.sin(x)

pi = math.pi
print('output:', taylor_poly(sin_equation, n_sin_equation, 6, pi/2, pi), '\n', 'x=6 n=pi')

print('output:', taylor_poly(sin_equation, n_sin_equation, 5, pi/2, pi/4), '\n', 'x=5 n=pi/4')
```

```python
''' Problem #4'''

def taylor_poly_bounded(func, n_func_deriv, c, x, epsilon):
    output = func(c)
    n = 1
    while True:
        term = n_func_deriv(c, n) * ((x - c) ** n) / math.factorial(n)
        if abs(term) < epsilon:
            break
        output += term
        n += 1
    return output

print('output:', taylor_poly_bounded(e_power, n_e_power, 0, 0.5, 0.1), '\n', 'x=0.5 epsilon=0.1')

print('output:', taylor_poly_bounded(e_power, n_e_power, 0, 0.25, 0.05), '\n', 'x=0.25 epsilon=0.05')

print('output:', taylor_poly_bounded(geo_equation, n_geo_equation, 0, 0.5, 0.1), '\n', 'x=0.5 epsilon=0.1')

print('output:', taylor_poly_bounded(geo_equation, n_geo_equation, 0, 0.25, 0.05), '\n', 'x=0.25 epsilon=0.05')

print('output:', taylor_poly_bounded(sin_equation, n_sin_equation, math.pi/2, math.pi/2 + 0.5, 0.1), '\n', 'x=0.5 epsilon=0.1')

print('output:', taylor_poly_bounded(sin_equation, n_sin_equation, math.pi/2, math.pi/2 + 0.25, 0.05), '\n', 'x=0.25 epsilon=0.05')
```
